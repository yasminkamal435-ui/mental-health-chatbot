```python
import gradio as gr
import pandas as pd
from sentence_transformers import SentenceTransformer, util

# قراءة البيانات
df = pd.read_csv("train.csv")
df["Context"] = df["Context"].fillna("")
df["Response"] = df["Response"].fillna("عذرًا، لا أملك إجابة على هذا السؤال.")

# تحميل موديل Sentence-BERT لتحويل الجمل إلى embeddings
model = SentenceTransformer('all-MiniLM-L6-v2')

# إنشاء embeddings لكل السياقات
embeddings = model.encode(df["Context"].tolist(), convert_to_tensor=True)

# دالة الشات بوت
def chatbot(user_input):
    if not user_input.strip():
        return "من فضلك اكتب سؤالاً."
    user_emb = model.encode(user_input, convert_to_tensor=True)
    # حساب التشابه
    cosine_scores = util.cos_sim(user_emb, embeddings)
    idx = cosine_scores.argmax().item()
    return df["Response"].iloc[idx]

# CSS لتحسين الخلفية
css = """
#chatbot {
  background: url("https://images.unsplash.com/photo-1507525428034-b723cf961d3e") no-repeat center center fixed;
  background-size: cover;
  border-radius: 10px;
  padding: 10px;
  color: white;
}
.gradio-chatbot-message {
    background-color: rgba(0,0,0,0.5) !important;
    color: white !important;
}
.gradio-textbox {
    border-radius: 10px;
}
"""

# إنشاء واجهة Gradio
with gr.Blocks(css=css) as demo:
    chatbot_ui = gr.Chatbot(elem_id="chatbot")
    msg = gr.Textbox(placeholder="اكتب سؤالك هنا...")

    def respond(message, chat_history):
        bot_message = chatbot(message)
        chat_history.append((message, bot_message))
        return "", chat_history

    msg.submit(respond, [msg, chatbot_ui], [msg, chatbot_ui])

demo.launch()
```

```python
import streamlit as st
import pandas as pd
from sentence_transformers import SentenceTransformer, util

# =======================
# تحميل البيانات
# =======================
df = pd.read_csv("train.csv")
df["Context"] = df["Context"].fillna("")
df["Response"] = df["Response"].fillna("عذرًا، لا أملك إجابة على هذا السؤال.")

# =======================
# تحميل موديل Sentence-BERT
# =======================
model = SentenceTransformer('all-MiniLM-L6-v2')
embeddings = model.encode(df["Context"].tolist(), convert_to_tensor=True)

# =======================
# دالة الشات بوت
# =======================
def chatbot(user_input):
    if not user_input.strip():
        return "من فضلك اكتب سؤالاً."
    user_emb = model.encode(user_input, convert_to_tensor=True)
    cosine_scores = util.cos_sim(user_emb, embeddings)
    idx = cosine_scores.argmax().item()
    return df["Response"].iloc[idx]

# =======================
# واجهة Streamlit
# =======================
st.set_page_config(page_title="ChatBot للصحة النفسية", layout="wide")

st.title("💬 شات بوت للصحة النفسية")
st.write("اكتب سؤالك وسأحاول الرد عليك بأفضل شكل ممكن.")

# تخزين المحادثة في الجلسة
if "chat_history" not in st.session_state:
    st.session_state.chat_history = []

# صندوق الرسائل
user_input = st.text_input("اكتب سؤالك هنا:")

if st.button("ارسال") and user_input:
    response = chatbot(user_input)
    st.session_state.chat_history.append((user_input, response))

# عرض المحادثة
for user_msg, bot_msg in st.session_state.chat_history:
    st.markdown(f"**أنت:** {user_msg}")
    st.markdown(f"**البوت:** {bot_msg}")
```
